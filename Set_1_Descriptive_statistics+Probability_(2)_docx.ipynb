```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
import statistics as stat
import warnings
warnings.filterwarnings('ignore')
```

```python
x=pd.Series([24.23,25.53,25.41,24.14,29.62,28.25,25.81,24.39,40.26,32.95,91.36,25.99,39.42,26.71,35.00])
```

```python
name=['Allied Signal','Bankers Trust','General Mills','ITT Industries','J.P.Morgan & Co.','Lehman Brothers',
      'Marriott','MCI','Merrill Lynch','Microsoft','Morgan Stanley','Sun Microsystems','Travelers','US Airways',
      'Warner-Lambert']
```

```python
# Pie Plot
plt.figure(figsize=(6,8))
plt.pie(x,labels=name,autopct='%1.0f%%')
plt.show()
```

```python
# Box Plot to find outliars
sns.boxplot(x)
```

```python
# barplot
sns.barplot(x=name,y=x)
plt.xticks(rotation=90)
plt.show()
```

```python
# Outliers
x = pd.Series(x)
q1 = np.quantile(x,0.25)
q3 = np.quantile(x,0.75)
q2 = np.median(x)
IQR = q3 - q1
upper_bond = q3+(1.5*IQR)
lower_bond = q1-(1.5*IQR)
outliers = x[(x <= lower_bond) | (x >= upper_bond)]
print('The Outlier in the boxplot:',outliers)
```

```python
print('Mean :',stat.mean(x))
print('Variance :',stat.variance(x))
print('Standard Deviation :',stat.stdev(x))
```
